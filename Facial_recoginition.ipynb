```python
import cv2 # ---> latest computer vision library
from google.colab.patches import cv2_imshow # this module help to show the image in google colab
# Specify the location of image and haar cascade files

cascadeFile = 'haarcascade_frontalface_default.xml'
imagePath = 'akshay.jpg'

# create the haar cascade object
# cascadeclassifier is used to specify which cascadefile will be used
faceCascade = cv2.CascadeClassifier(cascadeFile) #specify the cascade file location by passing the variable name

# read the image
#imread --> module --> refers to image read

image = cv2.imread(imagePath)

# convert image into grayscale to process the image more faster

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print(gray)
# BGR ---> Blue Green Red
# BGR is a standard color format for pictures taken by a camera

# detect face in an image 
# detectMultiScale() ---> this method is used to detect objects
# firstParameter ---> grayScale image
# secondParameter ---> scaleFactor # Parameter specifying how much the image size is reduced at each image scale. default = 1.3
# thirdParameter ---> minNeighbors # Specifying how many neighbors each candidate rectangle should have to retain it. (between 3-6) #affect quality of detected faces
# forthParameter ---> minSize # Minimum possible object size. Objects smaller than that are ignored. (30,30)

faces = faceCascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=9,minSize=(30,30)) #if scaleFactor=1.3, it will produce 2 rectangles
faces
# return 4 values as x,y,w,h in the next step

# draw rectangle around face
# cv2.rectangle(originalImage, startPoint, endPoint, color, thickness)  --> colour specify rgb colour code

for x,y,w,h in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)

# display the final output
cv2_imshow(image)
```
